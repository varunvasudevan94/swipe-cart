```python
import pandas as pd
import urllib
import os
import json

FILE_NAME = 'grocery.csv'
IMAGE_DST = 'img'
COLUMNS = ['asin', 'price.current_price', 'reviews.total_reviews', 'score', 'reviews.rating', 'bestSeller', 'thumbnail']
JSON_FILE = 'grocery.json'

from pathlib import Path
Path(IMAGE_DST).mkdir(parents=True, exist_ok=True)

df = pd.read_csv(FILE_NAME).set_index('asin')
```

```python
def get_unique_image_name(url):
    return url.split('/')[-1]
```

```python
def download(url):
    print(url)
    urllib.request.urlretrieve(url, os.path.join(IMAGE_DST, get_unique_image_name(url)))
```

```python
_ = df.thumbnail.map(download)
```

```python
df['image'] = df.thumbnail.map(lambda x: os.path.join('.', IMAGE_DST, get_unique_image_name(x)))
```

```python
rec_dict = df.T.to_dict()
```

```python
def dump_json(rec_dict):
    with open(JSON_FILE, 'w') as f:
        json.dump(rec_dict, f)
```

```python
dump_json(rec_dict)
```
